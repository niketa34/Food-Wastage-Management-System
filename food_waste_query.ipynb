import pandas as pd
import mysql.connector
import os

# List of CSV files and their corresponding table names
csv_files = [('claims_data.csv','claims_data'),('food_listings_data.csv','food_listings_data'),('providers_data.csv','providers_data'),('receivers_data.csv','receivers_data')  # Added payments.csv for specific handling
]

# Connect to the MySQL database
conn = mysql.connector.connect(
    host='localhost',
    user='root',
    password='admin@123',
    database='food_wastage_solutn'
)
cursor = conn.cursor()

# Folder containing the CSV files
folder_path = 'C:/Users\ADMIN/Documents/project_1data'

def get_sql_type(dtype):
    if pd.api.types.is_integer_dtype(dtype):
        return 'INT'
    elif pd.api.types.is_float_dtype(dtype):
        return 'FLOAT'
    elif pd.api.types.is_bool_dtype(dtype):
        return 'BOOLEAN'
    elif pd.api.types.is_datetime64_any_dtype(dtype):
        return 'DATETIME'
    else:
        return 'TEXT'

for csv_file, table_name in csv_files:
    file_path = os.path.join(folder_path, csv_file)
    
    # Read the CSV file into a pandas DataFrame
    df = pd.read_csv(file_path)
    
    # Replace NaN with None to handle SQL NULL
    df = df.where(pd.notnull(df), None)
    
    # Debugging: Check for NaN values
    print(f"Processing {csv_file}")
    print(f"NaN values before replacement:\n{df.isnull().sum()}\n")

    # Clean column names
    df.columns = [col.replace(' ', '_').replace('-', '_').replace('.', '_') for col in df.columns]

    # Generate the CREATE TABLE statement with appropriate data types
    columns = ', '.join([f'`{col}` {get_sql_type(df[col].dtype)}' for col in df.columns])
    create_table_query = f'CREATE TABLE IF NOT EXISTS `{table_name}` ({columns})'
    cursor.execute(create_table_query)

    # Insert DataFrame data into the MySQL table
    for _, row in df.iterrows():
        # Convert row to tuple and handle NaN/None explicitly
        values = tuple(None if pd.isna(x) else x for x in row)
        sql = f"INSERT INTO `{table_name}` ({', '.join(['`' + col + '`' for col in df.columns])}) VALUES ({', '.join(['%s'] * len(row))})"
        cursor.execute(sql, values)

    # Commit the transaction for the current CSV file
    conn.commit()

# Close the connection
conn.close()

1. query to write How many food providers and receivers are there in each city

```python
import pymysql
import pandas as pd

cur = connection.cursor()
query = """

            SELECT 
                l.Location,
                COUNT(p.Provider_ID) AS Provider_count,
                COUNT(r.Receiver_ID) AS Receiver_count
            FROM food_wastage_solutn.food_listings_data l
            LEFT JOIN providers_data p ON p.Provider_ID = l.Provider_ID
            LEFT JOIN claims_data c ON c.Food_ID = l.Food_ID
            LEFT JOIN receivers_data r ON r.Receiver_ID = c.Receiver_ID
            GROUP BY l.Location

        """
cur.execute(query)

# Step 4: Fetch and load into DataFrame
data = cur.fetchall()
df = pd.DataFrame(data, columns=["Location","Provider_count","Receiver_count"])
print(df)
```

4. query for Which receivers have claimed the most food?

```python
cur = connection.cursor()

# Step 3: Define and execute query
query = """
SELECT Receiver_ID, Name, claim_count
FROM (
    SELECT r.Receiver_ID, r.Name, COUNT(c.Claim_ID) AS claim_count,
           RANK() OVER (ORDER BY COUNT(c.Claim_ID) DESC) AS rank_val
    FROM receivers_data r
    JOIN claims_data c ON c.Receiver_ID = r.Receiver_ID
    GROUP BY r.Receiver_ID, r.Name
) AS ranked
WHERE rank_val = 1
ORDER BY Receiver_ID;
"""
cur.execute(query)

# Step 4: Fetch and load into DataFrame
data = cur.fetchall()
df = pd.DataFrame(data, columns=["Receiver_ID", "Name", "claim_count"])

# Step 5: Preview result
print(df)
```

6. query for Which city has the highest number of food listings?

```python
cur=connection.cursor()
query = """SELECT Location, foodlist_count
FROM (
    SELECT 
        Location,
        COUNT(*) AS foodlist_count,
        RANK() OVER (ORDER BY COUNT(*) DESC) AS rank_val
    FROM food_listings_data
    GROUP BY Location
) AS ranked_locations
WHERE rank_val = 1
ORDER BY Location;
"""
cur.execute(query)

data = cur.fetchall()
df = pd.DataFrame(data, columns=["Location","foodlist_count"])

print(df)
```

10. What percentage of food claims are completed vs. pending vs. canceled?

```python
cur=connection.cursor()

query = """
SELECT 
    Status,
    COUNT(*) AS claim_count,
    ROUND((COUNT(*) * 100.0 / (SELECT COUNT(*) FROM claims_data)), 2) AS percentage
FROM claims_data
GROUP BY Status;
"""
cur.execute(query)

data = cur.fetchall()
df = pd.DataFrame(data, columns=["Status", "claim_count","percentage"])

print(df.head())
```

14. Which food items are most frequently claimed before their expiry date?

```python
cur = connection.cursor()

query = """
SELECT fl.Food_Name,COUNT(c.Claim_ID) AS Claims_Before_Expiry
FROM claims_data c
JOIN food_listings_data fl ON c.Food_ID = fl.Food_ID
WHERE c.Timestamp < fl.Expiry_Date
GROUP BY fl.Food_Name
ORDER BY Claims_Before_Expiry DESC
LIMIT 10;
"""
cur.execute(query)

data = cur.fetchall()
df = pd.DataFrame(data, columns=["Food_Name","Claims_Before_Expiry"])

print(df)
```

15. Which cities have the highest ratio of claims to listings?

```python
cur = connection.cursor()

query = """ SELECT 
    fl.location,
    COUNT(DISTINCT c.Claim_ID) AS Total_Claims,
    COUNT(DISTINCT fl.Food_ID) AS Total_Listings,
    ROUND(COUNT(DISTINCT c.Claim_ID) / COUNT(DISTINCT fl.Food_ID), 2) AS Claim_to_Listing_Ratio
FROM 
    food_listings_data fl
LEFT JOIN 
    claims_data c ON fl.Food_ID = c.Food_ID
GROUP BY 
    fl.Location
HAVING 
    Total_Listings > 0
ORDER BY 
    Claim_to_Listing_Ratio DESC;
    """
cur.execute(query)

data = cur.fetchall()
df = pd.DataFrame(data, columns=["location","Total_Claims","Total_Listings","Claim_to_Listing_Ratio"])

print(df)
```

16. What is the distribution of food types (Vegetarian, Non-Vegetarian, Vegan) claimed by receiver type (NGO, Individual, etc.)?

```python
cur = connection.cursor()

query = """ SELECT 
    r.type as Receiver_Type,
    fl.Food_Type,
    COUNT(*) AS Total_Claims
FROM 
    claims_data c
JOIN 
    food_listings_data fl ON c.Food_ID = fl.Food_ID
JOIN 
    receivers_data r ON c.Receiver_ID = r.Receiver_ID
GROUP BY 
    r.type, fl.Food_Type
ORDER BY 
    r.type, Total_Claims DESC;
    """
cur.execute(query)

data = cur.fetchall()
df = pd.DataFrame(data, columns=["Receiver_Type","Food_Type","Total_Claims"])

print(df)
```

17. What is the distribution percentages of food types (Vegetarian, Non-Vegetarian, Vegan) claimed by each receiver type (NGO, Individual, etc.)

```python
cur = connection.cursor()

query = """ SELECT 
    r.Type as Receiver_Type,
    fl.Food_Type,
    COUNT(*) AS Total_Claims,
    ROUND(
        COUNT(*) * 100.0 / SUM(COUNT(*)) OVER (PARTITION BY r.Type), 2
    ) AS Percentage_Within_Receiver
FROM 
    claims_data c
JOIN 
    food_listings_data fl ON c.Food_ID = fl.Food_ID
JOIN 
    receivers_data r ON c.Receiver_ID = r.Receiver_ID
GROUP BY 
    r.Type, fl.Food_Type
ORDER BY 
    Percentage_Within_Receiver DESC;
    """
cur.execute(query)

data = cur.fetchall()
df = pd.DataFrame(data, columns=["Receiver_Type","Food_Type","Total_Claims","Percentage_Within_Receiver"])

print(df)
```

19. Which city have heighest number of food claims.

```python
cur = connection.cursor()

# Step 3: Define and execute query
query = """
SELECT city, claim_count
FROM (
    SELECT r.city,
           COUNT(c.Claim_ID) AS claim_count,
           RANK() OVER (ORDER BY COUNT(c.Claim_ID) DESC) AS rk
    FROM receivers_data r
    INNER JOIN claims_data c ON c.Receiver_ID = r.Receiver_ID
    GROUP BY r.city
) AS ranked
WHERE rk = 1;
"""

cur.execute(query)

# Step 4: Fetch and load into DataFrame
data = cur.fetchall()
df = pd.DataFrame(data, columns=["city", "claim_count"])

# Step 5: Preview result
print(df)
```
